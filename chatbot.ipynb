```python
def chatbot():
    responses = {
        "hello": "Hello! Welcome to our support. How can I help you?",
        "hi": "Hi there! How can I assist you today?",
        "order": "You can track your order at our website under 'My Orders'.",
        "refund": "Refunds are processed within 5-7 business days. Shall I raise a request?",
        "payment": "We accept Credit Card, Debit Card, UPI, and Net Banking.",
        "hours": "Our support is available Mon–Sat, 9 AM to 6 PM.",
        "bye": "Thank you for contacting us. Have a great day!",
        "help": "I can help with: order, refund, payment, hours."
    }

    print("Chatbot: Hello! I'm your customer support assistant.")
    print("Chatbot: Type 'bye' to exit.\n")

    while True:
        user_input = input("You: ").strip().lower()
        matched = False
        for key in responses:
            if key in user_input:
                print(f"Chatbot: {responses[key]}\n")
                matched = True
                break
        if not matched:
            print("Chatbot: Sorry, I didn't understand. Type 'help' to see options.\n")
        if "bye" in user_input:
            break

chatbot()
```
